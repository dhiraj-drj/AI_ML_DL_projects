```python
df.drop(["deck", "embark_town","alive", "who", "adult_male","class"], axis=1, inplace=True)   #droping all these columns as there is no such uses of these columns.
```

```python
df   #all those columns are removed from the dataset.
```

```python
df.isnull().sum()    #columns "age" and "embarked" have null values.
```

```python
df["age"].fillna(df["age"].mean(), inplace=True)   #filling the null rows in the column 'age' with 'mean'.
```

```python
df.dropna(subset=["embarked"],inplace=True)  #removing the 2 null rows in the column 'embarked' and then it will remove these 2 rows from all the columns.
```

```python
df.isna().sum()   #now there is no any null values.
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked']) #S=2,C=0,Q=1
```

```python
df
```

```python
df = df.astype(int)
```

```python
df
```

```python
x = df.drop("survived", axis=1)
y = df["survived"]
```

```python
x
```

```python
y
```

  Spliting of Dataset into Training and Testing Data.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Model Creation Usinh Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics  import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
