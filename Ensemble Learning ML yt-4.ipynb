```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df=sns.load_dataset("iris")
df
```

```python
df.head()
```

```python
x=df.drop("species",axis=1)
y=df["species"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
y_encoded=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

                    Training Model Using Stacking Learning.

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
base_learners=[
    ("dt",DecisionTreeClassifier(random_state=42)),
    ("lr",LogisticRegression(max_iter=1000,random_state=42)),
    ("SVC",SVC(probability=True,kernel='rbf',random_state=42))
]
```

```python
meta_learner=LogisticRegression(max_iter=1000,random_state=42)  #we can also use other Model such as KNN,...
```

```python
from sklearn.ensemble import StackingClassifier
```

```python
stacking_clf = StackingClassifier(
    estimators=base_learners,
    final_estimator=meta_learner,
    cv=5
)
```

```python
stacking_clf.fit(x_train,y_train)
```

```python
y_pred = stacking_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

                    Training Model Using Bagging Learning With Random Forest Concept
                    

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    random_state=42
)
```

```python
rf_model.fit(x_train,y_train)
```

```python
y_pred = rf_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

           Training Model Using Boosting Learning.
